```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# load dataset
# Read the csv file into a pandas DataFrame

nfl_df = pd.read_csv("resources/nfl_data_with_final_win_percentages.csv")
# list(nfl_df)
nfl_df.head()
```

```python
#clean dataframe

nfl_df = nfl_df.drop(columns=['Unnamed: 0','team','team_id','team_favorite_id','stadium_indoors_outdoors','weather_temperature',
                              'stadium','stadium_surface','stadium_indoors_outdoors','weather_humidity',
                              'stadium_neutral','weather_detail','weather_wind_mph','stadium_neutral','schedule_date',
                              'schedule_season','schedule_week'])
# Drop the null columns where all values are null
nfl_df = nfl_df.dropna(axis='columns', how='all')
# Drop the null rows
nfl_df = nfl_df.dropna()
nfl_df.head()
```

```python
#create train test split

from sklearn.model_selection import train_test_split
y = nfl_df['win_lose_draw']
X = nfl_df.drop(columns=['win_lose_draw'])

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
X_train.head()
```

```python
#pre-processing

from sklearn.preprocessing import LabelEncoder, MinMaxScaler
X_scaler = MinMaxScaler().fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
#Train the SVM
# Support vector machine linear classifier
from sklearn.svm import SVC 
model2 = SVC(kernel='linear')
model2.fit(X_train_scaled, y_train)
```

```python
SVC(C=1.0, cache_size=200, class_weight=None, coef0=0.0,
  decision_function_shape='ovr', degree=3, gamma='auto', kernel='linear',
  max_iter=-1, probability=False, random_state=None, shrinking=True,
  tol=0.001, verbose=False)
```

```python
print(f"Training Data Score: {model2.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model2.score(X_test_scaled, y_test)}")
```

```python
#Hyperparameter tuning

from sklearn.model_selection import GridSearchCV
param_grid = {'C': [1, 5, 10],
              'gamma': [0.0001, 0.001, 0.01]}
grid2 = GridSearchCV(model2, param_grid, verbose=3)
```

```python
#this takes time to run
grid2.fit(X_train_scaled, y_train)
```
