```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# load dataset
# Read the csv file into a pandas DataFrame

nfl_df = pd.read_csv("resources/nfl_data_with_final_win_percentages.csv")
list(nfl_df)
# nfl_df.head()
```

```python
#clean dataframe

nfl_df = nfl_df.drop(columns=['Unnamed: 0','team','team_id','team_favorite_id','stadium_indoors_outdoors','weather_temperature',
                              'stadium','stadium_surface','stadium_indoors_outdoors','weather_humidity',
                              'stadium_neutral','weather_detail','weather_wind_mph','stadium_neutral','schedule_date',
                              'schedule_season','schedule_week'])
# Drop the null columns where all values are null
nfl_df = nfl_df.dropna(axis='columns', how='all')
# Drop the null rows
nfl_df = nfl_df.dropna()
nfl_df.head()
```

```python
#create train test split
from sklearn.model_selection import train_test_split

y = nfl_df['win_lose_draw']
X = nfl_df.drop(columns=['win_lose_draw'])

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
X_train.head()
```

```python
#pre-processing
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

X_scaler = MinMaxScaler().fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.layers import Dense
```

```python
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)

y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
```

```python
 y_train_categorical.shape
```

```python
#create and train deep learning model using keras

model = Sequential()
model.add(Dense(units=100, activation='relu', input_dim=65))
model.add(Dense(units=100, activation='relu'))
model.add(Dense(units=3, activation='softmax'))
```

```python
# Compile and fit the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=60,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = model.evaluate(X_test_scaled, y_test_categorical, verbose=2)
print(f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
encoded_predictions = model.predict_classes(X_test_scaled[:5])
prediction_labels = label_encoder.inverse_transform(encoded_predictions)
```

```python
print(f"Predicted classes: {prediction_labels}")
print(f"Actual Labels: {list(y_test[:5])}")
```

```python
predictions = model.predict(X_test)
predictions
```

```python
# Save the model
model.save("nfl_keras_DL")
```

```python
# Load the model
from tensorflow.keras.models import load_model
model = load_model("nfl_keras_DL")
```

```python
# Evaluate the model using the training data
model_loss, model_accuracy = model.evaluate(X_test_scaled, y_test_categorical, verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
