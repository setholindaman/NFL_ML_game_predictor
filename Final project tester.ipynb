```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
games = pd.read_csv('resources/nfl_data_with_final_win_percentages.csv')
games.head()
```

```python
#games.drop("stadium_neutral", axis=1, inplace=True)
games = games.dropna(inplace=True)
games_d = games.astype(float)
games_clean.head()
```

```python
X = games.drop("win_lose_draw", axis=1)
y = games["win_lose_draw"].values.reshape(-1, 1)
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
# Create a logistic regression model

from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
classifier.score(X_train, y_train)
print(f'Clf score on train data is {classifier.score(X_train, y_train)}')
print(f'Clf score on test data is {classifier.score(X_test, y_test)}')
```

```python
predictions = classifier.predict(X_test)
```

```python
accuracy = pd.DataFrame({"Prediction": list(predictions), "Actual": y_test.ravel()}).reset_index(drop=True)
accuracy['correct'] = accuracy.Prediction == accuracy.Actual
accuracy.head(15)
```
